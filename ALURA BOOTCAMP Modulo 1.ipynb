```python
### Importando dados 
sel_encode= "ISO-8859-1"
pd.options.display.float_format='{:.2f}'.format
dados = pd.read_csv('A013051189_28_143_208.csv', sep=";", encoding = sel_encode, skipfooter=12, 
                    skiprows=3, thousands='.', decimal=',', engine='python')
dados = dados.replace('-', np.nan)
dados.head(3)
```

```python
dados = dados.sort_values(by=['2020/Ago'], ascending = False)
ax = dados.plot(x="Unidade da Federação", y= '2020/Ago', kind='bar', figsize=(10,7), fontsize=(12), color=['black'])
plt.title('Gastos por Unidade da Federação')
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.2f}'))
```
